### This script helps concatenating data files from multiple data folders under your local path
- Data path structures may look like this: Data folder--> Sub folder1/SF2/SF3... --> data files (see template)

```python
import pandas as pd
import re
import os
```

```python
#Read a file to create an empty dataframe for merging
path = '/Users/ze/Desktop/Template/' #replace with your local data path
data = pd.read_csv(path + 'Jan data/data_file_Jan_1.csv') #replace with a data file name/path; e.g., 'Jan data/data_file_Jan_1.csv'
data_merge= pd.DataFrame(columns = data.columns)
data_merge
```

```python
dayfiles = ['Jan data','Feb data' ]#replace with your local sub data folders; e.g., 'Jan data','Feb data' 
CommonName = 'data_file_' #replace with the common part in the target data file names; e.g., 'data_file_'

#Start to merge
length = 0

for dayfile in dayfiles:
    
    path = '/Users/ze/Desktop/Template/' + dayfile + '/'

    #Merge Data
    print( '---------------------------------------------------------')
    print('Folder: ', dayfile)
    
    for root, dirs, files in os.walk(path):
        
        for filename in files:
            
            if CommonName in filename and '._' not in filename:
                print('File: ', filename)
                    
                #read data
                data_new = pd.read_csv(path + filename)

                data_merge = pd.concat([data_merge, data_new], ignore_index=True, sort= True)

                length += len(data_new)

    print(length == len(data_merge))             
    print( 'Total Volume is: ', length)
```
